##Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]
```

```python
#Initialize max and min
maximum = numbers[0]
minimum = numbers[0]
```

```python
#loop through the numbers
for num in numbers:
  if num > maximum:
    maximum = num
  if num < minimum:
    minimum = num
```

```python
#output
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```
