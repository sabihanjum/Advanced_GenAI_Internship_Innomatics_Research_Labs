##2. Pass/Fail Analyzer

```python
#STudent marks
marks = [45, 78, 90, 33, 60]
```

```python
#Initialize counters
pass_count = 0
fail_count = 0
```

```python
#Analyze marks
for mark in marks:
  if mark >= 50:
    pass_count += 1
  else:
    fail_count += 1
```

```python
#print result
print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```
